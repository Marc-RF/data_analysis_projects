# Visualizing Earnings Based On College Majors

## Introduction

We are going to explore a dataset containing information about the job outcomes of students who graduated from college between 2010 and 2012.

The original dataset was released by the "American Community Survey" and cleaned by "FiveThirtyEight". It can be downloaded from their [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors) with the name of "recent-grads.csv". It also can be downloaded from the DataQuest guided project, by clicking on "Download" button over the embedded jupyter section.

## Dataset dictionary

Regarding the columns in the dataset, we are interested in:

- **Rank**: Rank by median earnings (the dataset is ordered by this column).
- **Major_code**: Major code.
- **Major**: Major description.
- **Major_category**: Category of major.
- **Total**: Total number of people with major.
- **Sample_size**: Sample size (unweighted) of full-time.
- **Men**: Male graduates.
- **Women**: Female graduates.
- **ShareWomen**: Women as share of total.
- **Employed**: Number employed.
- **Median**: Median salary of full-time, year-round workers.
- **Low_wage_jobs**: Number in low-wage service jobs.
- **Full_time**: Number employed 35 hours or more.
- **Part_time**: Number employed less than 35 hours.

Using visualizations, we can start to explore questions from the dataset like:

### Drawing scatter matrix plots

Pandas also offers the option to draw scatter matrix with multiple variables (columns) of a dataframe. The result is a matrix with histograms in the diagonal and scatter plots in the other cells.

The function scatter_matrix() is in pandas.plotting and must be imported separately.

```python
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(recent_grads[["Sample_size","Median"]], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[["Sample_size","Median","Unemployment_rate"]], figsize=(15,15))
```

### Drawing bar plots

Now it's turn for the bar plots. Bar plots can be created from series or from dataframes.
- If they are created from series, the x labels will be numerics (1..n). --> ex: Series.plot(kind="bar")
- If they are created from dataframes, we can choose the fields for the x axis. --> ex: DataFrame.plot.bar(x="Major", y="Women")

```python
#Top 10 majors and its women share
recent_grads[:10].plot.bar(x="Major",y="ShareWomen")
#Last 10 majors and its women share
recent_grads[-10:].plot.bar(x="Major",y="ShareWomen")
```

```python
#Top 10 majors and its unemployment rate
recent_grads[:10].plot.bar(x="Major",y="Unemployment_rate")
#Last 10 majors and its unemployment rate
recent_grads[-10:].plot.bar(x="Major",y="Unemployment_rate")
```
