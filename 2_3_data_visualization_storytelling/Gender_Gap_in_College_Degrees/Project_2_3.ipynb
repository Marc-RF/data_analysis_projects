#  Visualizing The Gender Gap In College Degrees

## Introduction

We are going to study the gender gap in some **STEM** (Science, Technology, Engineering and Mathematics) degrees across the years. The time period considered in the dataset starts from 1970 and finished in 2012.

The dataset includes the percentatge of women in 17 different degrees (each one in a column) plus an initial column with the year. It has  been downloaded from the DataQuest guided project but it originally comes from Randal Olson, a data scientist at University Pennsylvania, who cleaned the data and uploaded the final file in his [personal website](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv).

The goal of this project is to play with visualization tools in python (matplotlib) to get information from the database.

## Dataset dictionary

Below you can find the different fields in the dataset:

- **Year**
- **Agriculture**
- **Architecture**
- **Art and Performance**
- **Biology**
- **Business**
- **Communications and Journalism**
- **Computer Science**
- **Education**
- **Engineering**

- **Physical Sciences**
- **Psychology**
- **Public Administration**
- **Social Sciences and History**

## Initialization of data

Importing packages and defining variables:

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('my_datasets/percent_degrees_women/percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
```

## Plot created in previous exercises

Code from previous exercises to start visualizing the dataset:

```python
fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for _,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    # tick_params were assigned to "on"/"off" but it's recommended to use True/False
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

## Modifying y-axis ticks

Now it looks better but, as a counterpart, it is more difficult to discern the year when looking at the graphs. However, our goal is to visually check which degrees have bigger gaps.

Moreover, we can also reduce the y-axis ticks to only leave 0 and 100 positions.

```python
fig = plt.figure(figsize=(18, 18))

for sp in range(0,6):
    ax = fig.add_subplot(6,3,3*sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for _,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(stem_cats[sp])
    # tick_params were assigned to "on"/"off" but it's recommended to use True/False
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    
    if sp == 0:
        ax.text(2005, 10, 'Men')
        ax.text(2005, 85, 'Women')
    elif sp == 5:
        ax.text(2005, 90, 'Men')
        ax.text(2005, 10, 'Women')
        ax.tick_params(labelbottom=True)
        
for sp in range(0,5):
    ax = fig.add_subplot(6,3,3*sp+2)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for _,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(lib_arts_cats[sp])
    # tick_params were assigned to "on"/"off" but it's recommended to use True/False
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    
    if sp == 0:
        ax.text(2005, 20, 'Men')
        ax.text(2005, 80, 'Women')
    if sp == 4:
        ax.tick_params(labelbottom=True)

for sp in range(0,6):
    ax = fig.add_subplot(6,3,3*sp+3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for _,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(other_cats[sp])
    # tick_params were assigned to "on"/"off" but it's recommended to use True/False
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    
    if sp == 0:
        ax.text(2005, 5, 'Men')
        ax.text(2005, 90, 'Women')
    elif sp == 5:
        ax.text(2005, 65, 'Men')
        ax.text(2005, 35, 'Women')
        ax.tick_params(labelbottom=True)

plt.show()
```
